```python
arr = {}
for i in range(0, 100):
    nums = np.random.randint(0,100, size= 100)
    mean = np.mean(nums)
    if mean in arr:
        arr[mean] += 1
    else:
        arr[mean] = 1
```

```python
x = list(arr.keys())
y = list(arr.values())
```
